```python
m3.sort("CARD_ID", ascending=0)[["CARD_ID",u'NEWTOWN_BB',u'PLANNING_BB',u'REGION_BB',u'NEWTOWN_AA',u'PLANNING_AA',u'REGION_AA']]
```

```python
da2 = data_a .sort("Ride_Start_Date_Time", ascending = 0).groupby('CARD_ID').agg({
        'Ride_Start_Date_Time' :lambda x: x.iloc[-1],
        'Ride_End_Date_Time':lambda x: x.iloc[0],
        "REGION_AA" : lambda x: x.iloc[0],
        "PLANNING_AA" : lambda x: x.iloc[0],
        "NEWTOWN_AA" : lambda x: x.iloc[0],
        "REGION_BB" : lambda x: x.iloc[0],
        "PLANNING_BB" : lambda x: x.iloc[0],
        "NEWTOWN_BB" : lambda x: x.iloc[0],
        u'Ride_Distance':sum, 
        u'Ride_Time':sum, u'DIST':sum})

da2["Ride_Start_Date_Time"] = pd.to_datetime(da2["Ride_Start_Date_Time"])
da2["Ride_End_Date_Time"] = pd.to_datetime(da2["Ride_End_Date_Time"])
```

```python
da2["diff"] = abs(da2["Ride_Start_Date_Time"] - da2["Ride_End_Date_Time"])
da2["diff2"] =(abs(da2["diff"]) / pd.Timedelta( hours = 1 ))
```

```python
da3 = (da2[da2["diff2"]>4])
```

```python
da3.sort_index()[[  u'NEWTOWN_BB',u'PLANNING_BB',u'REGION_BB',u'NEWTOWN_AA',u'PLANNING_AA',u'REGION_AA']]
```

```python
da3["Ride_Start_Date_Time"] = pd.to_datetime(da3["Ride_Start_Date_Time"])
da3["hour"] = da3["Ride_Start_Date_Time"].apply(lambda x: x.hour)
```
